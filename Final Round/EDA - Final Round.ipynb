```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.seasonal import seasonal_decompose
```

### The morning and evening dataset used in this EDA session was obtained from concatinating result from '10 simpang'

```python
morn_june = pd.read_csv('/here-maps-dataset/df_morning_june.csv').drop(columns='Unnamed: 0', axis=1)
morn_june.head()
```

```python
# Decompose the travel_time_seconds column to get the trend, seasonality, and noise
# period 7 artinya kita punya data harian, dan pengen observe weekly trendnya

decompose_jarak = seasonal_decompose(morn_june.travel_time_seconds, model='multiplicative', period=7)
fig = decompose_jarak.plot()
#fig.savefig('decompose_travel time_pagi.png')
fig.show()
```

```python
eve_jun = pd.read_csv('/here-maps-dataset/df_evening_june.csv').drop(columns='Unnamed: 0', axis=1)
eve_jun.head()
```

```python
# Decompose the travel_time_seconds column to get the trend, seasonality, and noise
# period 7 artinya kita punya data harian, dan pengen observe weekly trendnya

decompose_jarak = seasonal_decompose(eve_jun.travel_time_seconds, model='multiplicative', period=7)
fig_eve = decompose_jarak.plot()
#fig_eve.savefig('decompose_travel time_sore.png')
fig_eve.show()
```

```python
#create variable to split up pagi dan sore for EDA purpose

def morn_evening(hour): #return int
    if hour <= 10:
        return 0
    else:
        return 1
    
def morn_evening_string(morn_evening): #return string
    if morn_evening == 0:
        return 'pagi'
    else:
        return 'sore'
```

```python
june_23['morning_evening'] = june_23['hour'].apply(morn_evening)
june_23['morning_evening_string'] = june_23['morning_evening'].apply(morn_evening_string)
```

```python
june_23.tail()
```

```python
simpang_waktu_cat = june_23.groupby(['simpang', 'morning_evening_string']).travel_time_seconds.agg('mean')
simpang_waktu_cat
```

```python
#save aggregation into new data frame
quest_3 = simpang_waktu_cat.reset_index()
quest_3.rename(columns={'simpang': 'Simpang jalan',
                        'morning_evening_string': 'Pagi / Sore',
                        'travel_time_seconds': 'Rerata waktu tempuh'}, inplace=True)
quest_3
```

```python
#visualize rerata waktu tempuh per simpang
sns.set()
fig = plt.figure(figsize = (16,12))
g = sns.barplot(data=quest_3,
                x="Simpang jalan",
                y="Rerata waktu tempuh",
                hue="Pagi / Sore",
                errorbar="sd")

#g.bar_label(g.containers[0])
plt.title('Rerata waktu tempuh per simpang', fontsize = 24, loc="left", color="#000", pad=5, fontweight="bold", fontfamily="serif", y=1.05, zorder=3)
plt.xlabel('Simpang', fontsize = 12)
plt.ylabel('Rerata waktu tempuh', fontsize = 18)
plt.xticks(fontsize = 15, rotation=45)
plt.yticks(fontsize = 15)
plt.show()
```

## Multivariate Analysis

```python
# corr matrix
yuk_cek = june_23[['travel_time_seconds', 'base_duration', 'length', 'speed', 'day', 'hour', 'minute']]

corr = yuk_cek.corr()
plt.figure(figsize=(12,8))
#masking the upper triangle part (since matrix is repetitive)
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap="Oranges", annot=True)
plt.savefig('korelasi fitur, label, fitur eng.png')
plt.show()
```
